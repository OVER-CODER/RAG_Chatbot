```python
from langchain_community.document_loaders import PyPDFLoader
import os

my_dir = "/content/books"
for f in os.listdir(my_dir):
  print(f"\n(f)")
  loader = PyPDFLoader(os.path.join(my_dir, f))
  doc = loader.load()
  print(type(doc), len(doc))
  print(type(doc[0]))
  print(doc[0])
```

### **Chunking**

```python
!pip install -qU langchain-text-splitters
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=500)
chunks = splitter.split_documents(doc)
print(f"{len(chunks)} documents have been split into {len(chunks)} chunks")
print(chunks[0].page_content)

```

### **Embeddings & Vectorstore**

```python
!pip install sentence-transformers
```

```python
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
!pip install -qU langchain-chroma
```

```python
from langchain_chroma import Chroma
persist_directory = os.path.join(my_dir, "chroma_db")

vectordb = Chroma.from_documents(
    documents = chunks,
    persist_directory=persist_directory,
    embedding=embeddings
)
```

### **Retrieval**


```python
query = "what is name of the chapter"
relevant_docs = vectordb.max_marginal_relevance_search(query)
print(f"Found {len(relevant_docs)} relevant documents")
print(relevant_docs[0].page_content)
```

```python
for d in relevant_docs:
  print(f"-----------{len(d.page_content)}-------\n{d.page_content}" )
```
